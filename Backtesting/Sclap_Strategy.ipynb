5 minute time frame

VWAP: 15 candles above or below for trend detection 

Bollinger bands strategy (length 14 std 2)

RSI for confirmation (RSI<45 buy , RSI>55 sell signal)

RSI for stop(RSI<10 get out of sell , RSI>90 get out of buy)

SL = a * ATR

TP = RRR * SL

```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import timedelta, datetime
import pandas_ta as ta
import matplotlib.pyplot as plt
import warnings
plt.style.use("seaborn-v0_8")
warnings.filterwarnings('ignore')
```

```python
ticker= "EURUSD=X"
ticker="AAPL"
bollinger_period= 14
rsi_period = 16
df=yf.download(ticker,period='60d', interval="5m")
df
```

```python
df['VWAP'] = ta.vwap(df['High'], df['Low'], df['Close'], df['Volume'])

df['RSI']=ta.rsi(df.Close, length=rsi_period)

df['sma'] = ta.sma(df['Close'], length=bollinger_period)
df['std'] = df['Close'].rolling(window=bollinger_period).std()
df['upper_band'] = df['sma'].add(df['std'].mul(2))
df['lower_band'] = df['sma'].add(df['std'].mul(-2))

df.dropna(inplace=True)
df
```

```python
import plotly.graph_objects as go
from datetime import datetime
dfpl = df.last('2D')
dfpl.reset_index(inplace=True)
fig = go.Figure(data=[go.Candlestick(x=dfpl.index,
                open=dfpl['Open'],
                high=dfpl['High'],
                low=dfpl['Low'],
                close=dfpl['Close']),
                go.Scatter(x=dfpl.index, y=dfpl.VWAP, 
                           line=dict(color='blue', width=1), 
                           name="VWAP"), 
                go.Scatter(x=dfpl.index, y=dfpl["lower_band"], 
                           line=dict(color='green', width=1), 
                           name="lower_band"),
                go.Scatter(x=dfpl.index, y=dfpl['upper_band'], 
                           line=dict(color='green', width=1),
                           fill='tonexty',  # Fill area between the bands
                           fillcolor='rgba(204, 204, 204, 0.5)',  # Grey color with some transparency
                           name="upper_band")])

# fig.add_scatter(x=dfpl.index, y=dfpl['pointposbreak'], mode="markers",
#                 marker=dict(size=10, color="MediumPurple"),
#                 name="Signal")

fig.update_layout(
    width=1000, 
    height=800,
    font=dict(size=20)
)
fig.show()
```

```python
df['candles_above_vwap'] = 0
df['candles_below_vwap'] = 0


for i in range(1, len(df)):
    if max(df['Close'].iloc[i],df['Open'].iloc[i]  )> df['VWAP'].iloc[i]:
        df['candles_above_vwap'].iloc[i] = df['candles_above_vwap'].iloc[i-1] + 1
        df['candles_below_vwap'].iloc[i] = 0
    elif min(df['Close'].iloc[i],df['Open'].iloc[i]) < df['VWAP'].iloc[i]:
        df['candles_below_vwap'].iloc[i] = df['candles_below_vwap'].iloc[i-1] + 1
        df['candles_above_vwap'].iloc[i] = 0

backcandles = 15        
#0 : no signal / 1 : sell / 2: buy       
df['Signal'] = 0
for i in range(1, len(df)):
    if (df['candles_above_vwap'].iloc[i] >=backcandles and df.Close[i]<=df['lower_band'][i] and df.RSI[i]<45):
        df['Signal'].iloc[i] = 2
    elif (df['candles_below_vwap'].iloc[i] >=backcandles and df.Close[i]>=df['upper_band'][i] and df.RSI[i]>55):
        df['Signal'].iloc[i] = 1
    else:
        df['Signal'].iloc[i] = 0
        
df
```

```python
ATR_period=7

df['ATR']=ta.atr(df.High, df.Low, df.Close, length=ATR_period)

def SIGNAL():
    return df.Signal


df.set_index("Datetime", inplace=True)
df
```

```python
from backtesting import Strategy
from backtesting import Backtest
a=1.2
RRR=1.5
class MyStrat(Strategy):
    initsize = 0.99
    mysize = initsize
    def init(self):
        super().init()
        self.signal1 = self.I(SIGNAL)

    def next(self):
        super().next()
        slatr = a*self.data.ATR[-1]
        TPSLRatio = RRR

        if len(self.trades)>0:
            if self.trades[-1].is_long and self.data.RSI[-1]>=90:
                self.trades[-1].close()
            elif self.trades[-1].is_short and self.data.RSI[-1]<=10:
                self.trades[-1].close()
        
        if self.signal1==2 and len(self.trades)==0:
            sl1 = self.data.Close[-1] - slatr
            tp1 = self.data.Close[-1] + slatr*TPSLRatio
            self.buy(sl=sl1, tp=tp1, size=self.mysize)
        
        elif self.signal1==1 and len(self.trades)==0:         
            sl1 = self.data.Close[-1] + slatr
            tp1 = self.data.Close[-1] - slatr*TPSLRatio
            self.sell(sl=sl1, tp=tp1, size=self.mysize)

bt = Backtest(dfpl, MyStrat, cash=100, margin=1/10, commission=0.00)
stat = bt.run()
stat['Win Rate [%]']
stat
```
